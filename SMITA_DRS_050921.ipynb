```python
Rolling dice simulator.
This project aims at comparing two values, one being an user input value and the other a random value generated by the 
dice simulator.
For the random number generation a library random has been imported and the method random.randint
with values from 1 to 6 is called, since a dice has only these values. 
This number is stored in the variable rolled_num
The user input is stored in another variable called selected.
If the user input is not between 1 and 6 message invalid choice is displayed.
Every time the user gives an input and the dice is rolled the values are compared and print statement generated accordingly

In the second code , tried to use  a user defined function choice(x,y), 
with two parameters passed. The code works fine without user defined function but only created for resuability

Source Codes: Google search through various sites.
```
